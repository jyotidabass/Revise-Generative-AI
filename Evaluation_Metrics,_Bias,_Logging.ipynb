# **ðŸ§ª Meaningful Mini Code (Word Overlap Logic)**

```python
reference = "AI will change healthcare significantly"
generated = "Artificial intelligence will transform healthcare"

ref_words = set(reference.lower().split())
gen_words = set(generated.lower().split())

overlap = ref_words.intersection(gen_words)

print("Overlapping words:", overlap)
print("Overlap score:", len(overlap) / len(ref_words))
```
